```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Sales': [100, 120, 90, 150, 180]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create Line Chart
sns.lineplot(x='Year', y='Sales', data=df, marker='o')

# Customize chart
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [120, 150, 90, 180]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create Bar Chart
sns.barplot(x='Product', y='Sales', data=df, color='red')

# Customize chart
plt.title('Bar chart ex.')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid(axis='y')
plt.tight_layout()

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = {
    'Marks': [45, 56, 67, 45, 78, 89, 67, 56, 90, 76, 55, 67, 85, 95, 73]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create Histogram
sns.histplot(data=df, x='Marks', bins=8, kde=True, color='yellow')

# Customize chart
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample Data
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [250, 150, 100, 200]

# Create Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(sales, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution pie chart')
plt.tight_layout()

# Show chart
plt.show()
```
